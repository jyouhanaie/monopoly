```python
def chance_card_action(chance_cards, card_pointer, board_position):
    card_says = chance_cards[card_pointer-1]
    
    if card_says == "advance to go": 
        return 1
    
    elif card_says == "go to illinois": 
        return 25
    
    elif card_says == "go to st charles": 
        return 12
    
    elif card_says == "go to nearest utility":
        if 13 < board_position < 29:
            board_position = 29
        else:
            board_position = 13
            
        return board_position
    
    elif card_says == "go to nearest railroad 1" or card_says == "go to nearest railroad 2":
        # railroads are at 6, 16, 26, 36
        if board_position > 6 and board_position < 16:
            board_position = 16
        elif board_position > 16 and board_position < 26:
            board_position = 26
        elif board_position > 26 and board_position < 36:
            board_position = 36
        else :
            board_position = 6
        
        return board_position
    
    elif card_says == "go back 3 spaces": 
        board_position -= 3
        if board_position < 1:
            board_position += 40
            
        return board_position
    
    elif card_says == "go to jail":
        return 11
    
    elif card_says == "go to reading railroad":
        return 6
    
    elif card_says == "go to boardwalk":
        return 40
    
    else:
        return board_position
```

```python
def chest_card_action(chest_cards, card_pointer, board_position):
    card_says = chest_cards[card_pointer-1]
    
    if card_says == "advance to go":
        return 1
    
    elif card_says == "go to jail":
        return 11
    
    else:
        return board_position
```
