```python

class Space():
    def __init__(self, name, position):
        self.name = name
        self.position = position
        
    def get_position():
        return self.position
        
class Property(Space):
    def __init__(self, color, cost, *args, **kwargs):
        
        self.color = color
        self.cost = cost
        
        self.house_hotel_count = 0
        
        super().__init__(*args, **kwargs)
        
        self.owner = "unowned"
        self.mortgaged = False
    
    def get_color():
        return self.color
    
    def get_cost():
        return self.cost
    
    def get_owner():
        return self.owner
    
    def get_status():
        return self.mortgaged

    
    def get_house_hotels():
        return self.house_hotel_count

    def get_rent_cost():
        return rent_costs[self.get_house_hotels()]
        
        
        
class Railroad(Space):
    def __init__(self, cost, *args, **kwargs):
        self.cost = cost
        super().__init__(*args, **kwargs)
        
        
class Utility(Property):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        

class Chance(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
        
class Community_Chest(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
    
class Jail(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
        
class Go_FreeParking_IncomeTax(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
        
class Go_to_Jail(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        

class Jail_or_Just_Visiting(Space):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
```

```python
class Player(pygame.sprite.Sprite):
    
    def __init__(self):
        super().__init__() 
        self.surf = pygame.Surface((20, 20))
        self.surf.fill((255,255,255))
        self.rect = self.surf.get_rect(center = (770, 770))
        
        self.pos = vec((10, 385))
        self.vel = vec(0,0)
        self.acc = vec(0,0)
        
        
        self.cash = 1500
        self.position = 1
        self.properties = []
        
        self.railroad_count = 0
        self.utility_count = 0
        
        self.getoutofjailfree = 0
        self.rolledDoubled = False
        self.doublesCounter = 0
        self.injail = False
        self.jailCounter = 0
        
        
        
    def get_position(self):
        return self.position
    
    def add_position(self, addValue):
        self.position = self.position + addValue;
        if self.position > 40:
            self.position = self.position - 40
        if self.position <1:
            self.position = self.position + 40
    
    def move(self, coordinates):
        self.pos.x = coordinates[0]
        self.pos.y = coordinates[1]
        self.rect.midbottom = self.pos
    
    def position_to_coordinates(self, position):
        for key in space_dict:
            if key == position:
                return space_dict[key]
    def list_properties(self):
        property_list = [propertyName.name for propertyName in self.properties]
        return property_list
P1 = Player()
P1.surf.fill((0,0,0))
```

```python
all_sprites = pygame.sprite.Group()

all_sprites.add(roll)
all_sprites.add(P1)

clicked = False

def playerTurn(player, rolls):
    all_sprites.remove(roll)
    if player.injail:
        player.jailCounter -= 1
        print("in jail")
        if player.jailCounter == 1:
            player.injail = False
            #implement player lose 50$
    else:
        if rolls[0] == rolls[1]:
            player.doublesCounter += 1
            print("doubles!")
            player.rolledDoubles = True
        else:
            player.rolledDoubles = False
            player.doublesCounter = 0
        if player.doublesCounter >= 3:
            player.move(player.position_to_coordinates(11))
            player.position = 11
            player.injail = True
            player.jailCounter = 3
            player.rolledDoubles = False
            player.doublesCounter = 0
        else:
            roll_sum = sum(rolls)
            player.add_position(roll_sum)
            player.move(player.position_to_coordinates(player.position))
            if player.position in property_positions:
                player.properties.append(property_objects[player.position])

    all_sprites.add(roll)


players = [P1]
playerCount = len(players)
turnCounter = 0;
while True:
    window.blit(bg_img,(0,0))
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            break
            
    #mouse coords
    mousePos = pygame.mouse.get_pos()
    
    #if hit button
    if 200 <= mousePos[0] <= 400 and 450 <= mousePos[1] <= 550: 
        if pygame.mouse.get_pressed()[0] == True:
            clicked = True
        if clicked:
            if pygame.mouse.get_pressed()[0] == False:
                clicked = False
                rolls = [roll.dice_roll(), roll.dice_roll()]  
                print (rolls)
                playerTurn(players[turnCounter], rolls)
                print(players[turnCounter].list_properties())
                if players[turnCounter].rolledDoubles:
                    turnCounter -= 1
                
                turnCounter += 1
                if turnCounter >= playerCount:
                    turnCounter = 0
                
    
    for entity in all_sprites:
        window.blit(entity.surf, entity.rect)
 
    pygame.display.update()
```
