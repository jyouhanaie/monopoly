```python
def chance_card_action(chance_cards, card_pointer, board_position):
    card_says = chance_cards[card_pointer-1]
    
    if card_says == "advance to go": 
        return 1
    
    elif card_says == "go to illinois": 
        return 25
    
    elif card_says == "go to st charles": 
        return 12
    
    elif card_says == "go to nearest utility":
        if 13 < board_position < 29:
            board_position = 29
        else:
            board_position = 13
            
        return board_position
    
    elif card_says == "go to nearest railroad 1" or card_says == "go to nearest railroad 2":
        # railroads are at 6, 16, 26, 36
        if board_position > 6 and board_position < 16:
            board_position = 16
        elif board_position > 16 and board_position < 26:
            board_position = 26
        elif board_position > 26 and board_position < 36:
            board_position = 36
        else :
            board_position = 6
        
        return board_position
    
    elif card_says == "go back 3 spaces": 
        board_position -= 3
        if board_position < 1:
            board_position += 40
            
        return board_position
    
    elif card_says == "go to jail":
        return 11
    
    elif card_says == "go to reading railroad":
        return 6
    
    elif card_says == "go to boardwalk":
        return 40
    
    else:
        return board_position
```

```python
def chest_card_action(chest_cards, card_pointer, board_position):
    card_says = chest_cards[card_pointer-1]
    
    if card_says == "advance to go":
        return 1
    
    elif card_says == "go to jail":
        return 11
    
    else:
        return board_position
```

```python
def increment_card_pointer(card_pointer):
    # based on 16 cards in each deck
    if card_pointer < 16:
        card_pointer += 1
        
    else:
        card_pointer = 1
    
    return card_pointer
```

```python
# START OF NEW / ORIGINAL CODE

class Space():
    def __init__(self, name, position):
        
        self.position = position
        
    def get_position:
        return self.position
        
class Property(Space):
    def __innit__(self, color, cost, *args, **kwargs):
        
        self.color = color
        self.cost = cost
        
        super().__init__(*args, **kwargs)
        
        self.owner = "unowned"
        self.mortgaged = False
    
    def get_color():
        return self.color
    
    def get_cost():
        return self.cost
    
    def get_owner():
        return self.owner
    
    def get_status():
        return self.mortgaged

        
class Colored_Property(Property):
    def __innit__(self, color, cost, rent_costs, *args, **kwargs):
        
        self.rents_costs= rent_costs
        
        super().__init__(*args, **kwargs)
        
        self.house_hotel_count = 0
        
        def get_house_hotels():
            return self.house_hotel_count
        
        def get_rent_cost():
            return rent_costs[self.get_house_hotels()]
        
        
class Railroad(Property):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        
        
class Utility(Property):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        

class Chance(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        
        
class Community_Chest(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        
    
class Jail(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        
        
class Go_FreeParking_IncomeTax(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        
        
class Go_to_Jail(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
        

class Jail_or_Just_Visiting(Space):
    def __init__(self, *args, **kwargs):
        super().__innit(*args, **kwargs)
```
